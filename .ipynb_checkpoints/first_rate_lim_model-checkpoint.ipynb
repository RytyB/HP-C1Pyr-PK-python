<h1> Rate-limited model testing </h1>
<p> Context: Let's assume the ML portion of the algorithm works and I can identify the rate-limiting step. Can I incorporate that knowledge into a model that can recover the numerical value of the rate constant of the rate limiting step? <br>
<li>Generate data using a set of pet PK parameters (kecp limited)</li>
<li>Fit modified version of P2L2 to noiseless curves</li>
<li>Recover kecp</li>
<li>Fit modified version of P2L2 to noisy curves</li>
<li>Recover kecp</li>
<li>Repeat for all possible rate limiting steps</li>

```python
import pkhpc1pyr as pk
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data from pet PK parameters
acq = pk.acq_params()
pk_kecp_rl_3pc = pk.pk_params(acq, 
    kve = 5,
    kecp = 0.2,
    kpl = 5,
    kecl = 
)
noiseless_data = pk.Sim_Data(pk_kecp_rl_3pc, acq, model=pk.P2L3)

vb = pk_kecp_rl_3pc.vb
vef = pk_kecp_rl_3pc.vef
ve = (1-vb)*vef
vc = 1-vb-ve

pk_kecp_rl_2pc = pk.pk_params(acq,
    kve = 5,
    kpl = 0.2 / ve,
)
rl_mod = pk.rl_kecp(parms=pk_kecp_rl_2pc, acq=acq)

# Plot results
fig, ax = plt.subplots(figsize=(6,3))

ax.scatter(acq.taxis, noiseless_data.pyrSig, color='k', marker='x', linewidth=0.25, label='[Pyr] Ground Truth')
ax.scatter(acq.taxis, noiseless_data.lacSig, color='r', marker='x', linewidth=0.25, label='[Lac] Ground Truth')
ax.plot(acq.taxis, rl_mod.pyrSig, color='k', label='[Pyr] RL Model')
ax.plot(acq.taxis, rl_mod.lacSig, color='r', label='[Lac] RL Model')

ax.set_xlabel('Time (s)', fontsize='xx-large')
ax.set_ylabel(r'$M_{xy}$', fontsize='xx-large')
ax.legend()

# print(f'Driving kecp = {pk_kecp_rl.kecp}\nFitted kecp = {rl_fit.popt.kpl}')
pk_kecp_rl_2pc.pretty_print()
plt.show()
```

```python
import pkhpc1pyr as pk
import numpy as np
import matplotlib.pyplot as plt

show_3_comp = True

# Generate synthetic data from pet PK parameters
acq = pk.acq_params()
pk_kecp_rl = pk.pk_params(acq, 
    kve = .2,  # <--- This needs to not be infinite. Why? Who knows!
    kecp = 0.05,
    kpl = 100,
    kecl = 100
)
vb = pk_kecp_rl.vb
vef = pk_kecp_rl.vef
ve = (1-vb)*vef
vc = 1-vb-ve
noiseless_data = pk.Sim_Data(pk_kecp_rl, acq, model=pk.P2L3)

fit_vars = ['kecp', 'VIFScale', 'kve', 'T1lac']
fit_parms = pk.fit_params(nfits=3, fit_vars=fit_vars)
rl_fit = pk.Fit(fit_parms=fit_parms, default_parms=pk_kecp_rl, acq=acq, inp_data=noiseless_data, model=pk.rl_kecp)

# Plot results
fig, ax = plt.subplots(1,2, figsize=(12,3))

ax[0].scatter(acq.taxis, noiseless_data.pyrSig, color='k', marker='x', linewidth=0.25, label='[Pyr] Ground Truth')
ax[0].scatter(acq.taxis, noiseless_data.lacSig, color='r', marker='x', linewidth=0.25, label='[Lac] Ground Truth')
ax[0].plot(acq.taxis, rl_fit.pyrSig, color='k', label='[Pyr] RL Model')
ax[0].plot(acq.taxis, rl_fit.lacSig, color='r', label='[Lac] RL Model')

ax[0].set_xlabel('Time (s)', fontsize='x-large')
ax[0].set_ylabel(r'$M_{xy}$', fontsize='x-large')
ax[0].legend()
ax[0].set_title('Fitted parameters', fontsize='xx-large')

# Secondary plot to play around with
pk_play = pk.pk_params(acq,
                      kve=.2,
                      kecp=0.05
                      )
rl_data = pk.rl_kecp(pk_play, acq)

ax[1].scatter(acq.taxis, noiseless_data.pyrSig, color='k', marker='x', linewidth=0.25, label='[Pyr] Ground Truth')
ax[1].scatter(acq.taxis, noiseless_data.lacSig, color='r', marker='x', linewidth=0.25, label='[Lac] Ground Truth')
ax[1].plot(acq.taxis, rl_data.pyrSig, color='k', label='[Pyr] RL Model')
ax[1].plot(acq.taxis, rl_data.lacSig, color='r', label='[Lac] RL Model')

ax[1].set_xlabel('Time (s)', fontsize='x-large')
ax[1].set_ylabel(r'$M_{xy}$', fontsize='x-large')
ax[1].legend()
ax[1].set_title('Manual parameters', fontsize='xx-large')


print(f'Driving kecp = {pk_kecp_rl.kecp}\nFitted kecp = {rl_fit.popt.kecp:.5f}')
perc_error = np.abs(pk_kecp_rl.kecp - rl_fit.popt.kecp) * 100 / pk_kecp_rl.kecp
print(f'Rate limiting step estimation error: {perc_error:.2f} %\n')

print(f'Driving kve = {pk_kecp_rl.kve}\nFitted kve = {rl_fit.popt.kve:.5f}')
kve_error = np.abs(pk_kecp_rl.kve - rl_fit.popt.kve) * 100 / pk_kecp_rl.kve
print(f'Extravasation estimation error: {kve_error:.2f} %\n')

print(f'Fit NMSE = {rl_fit.calculate_NMSE()[2] * 100:.3f} %')

rl_fit.popt.pretty_print()
plt.show()

if show_3_comp:
    fig,ax = plt.subplots(1,3, figsize=(18,3))
    pc_3_data = pk.P2L3(pk_kecp_rl, acq)

    ylim = np.max(pc_3_data.Mxy)
    
    ax[0].plot(acq.taxis, pc_3_data.Mxy[0], color='k')
    ax[0].set_title('Vascular', fontsize='x-large')
    ax[0].set_ylim( [0,ylim] )

    ax[1].plot(acq.taxis, pc_3_data.Mxy[2], color='k')
    ax[1].plot(acq.taxis, pc_3_data.Mxy[3], color='r')
    ax[1].set_title('Extracellular', fontsize='x-large')
    ax[1].set_ylim( [0,ylim] )

    ax[2].plot(acq.taxis, pc_3_data.Mxy[4], color='k')
    ax[2].plot(acq.taxis, pc_3_data.Mxy[5], color='r')
    ax[2].set_title('Intracellular', fontsize='x-large')
    ax[2].set_ylim( [0,ylim] )
    plt.show()
```

<h3> Adding in noise to the fits </h3>

```python
import pkhpc1pyr as pk
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data from pet PK parameters
acq = pk.acq_params()
pk_kecp_rl = pk.pk_params(acq, 
    kve = 0.2,  # <--- This needs to not be infinite. Why? Who knows!
    kecp = 0.05,
    kpl = 100,
    kecl = 100
)
vb = pk_kecp_rl.vb
vef = pk_kecp_rl.vef
ve = (1-vb)*vef
vc = 1-vb-ve
noiseless_data = pk.Sim_Data(pk_kecp_rl, acq, model=pk.P2L3)
synth_data = noiseless_data.noisify(peakSnr=20)

fit_vars = ['kecp', 'VIFScale', 'kve']
fit_parms = pk.fit_params(nfits=3, fit_vars=fit_vars)
rl_fit = pk.Fit(fit_parms=fit_parms, default_parms=pk_kecp_rl, acq=acq, inp_data=synth_data, model=pk.rl_kecp)

# Plot results
fig, ax = plt.subplots(figsize=(6,3))

ax.scatter(acq.taxis, synth_data.pyrSig, color='k', marker='x', linewidth=0.25, label='[Pyr] Ground Truth')
ax.scatter(acq.taxis, synth_data.lacSig, color='r', marker='x', linewidth=0.25, label='[Lac] Ground Truth')
ax.plot(acq.taxis, rl_fit.pyrSig, color='k', label='[Pyr] RL Model')
ax.plot(acq.taxis, rl_fit.lacSig, color='r', label='[Lac] RL Model')

ax.set_xlabel('Time (s)', fontsize='x-large')
ax.set_ylabel(r'$M_{xy}$', fontsize='x-large')
ax.legend()
ax.set_title('Fitted parameters', fontsize='xx-large')

print(f'Driving kecp = {pk_kecp_rl.kecp}\nFitted kecp = {rl_fit.popt.kecp:.5f}')
perc_error = np.abs(pk_kecp_rl.kecp - rl_fit.popt.kecp) * 100 / pk_kecp_rl.kecp
print(f'Rate limiting step estimation error: {perc_error:.2f} %\n')

print(f'Driving kve = {pk_kecp_rl.kve}\nFitted kve = {rl_fit.popt.kve:.5f}')
kve_error = np.abs(pk_kecp_rl.kve - rl_fit.popt.kve) * 100 / pk_kecp_rl.kve
print(f'Extravasation estimation error: {kve_error:.2f} %')

rl_fit.popt.pretty_print()

plt.show()

```
